```python
import pandas as pd

df = pd.read_csv('customer_data.csv')
```

```python
df['age'].agg(
    max_age = 'max',
    min_age = 'min',
    avg_age = 'mean')
```

```python
 df['full_name']=df['prefix'].astype(str)+" "+df['first'].astype(str)+" "+df['last'].astype(str)
top_names = df['full_name'].value_counts()
print(top_names)
```

```python
grouped = df.groupby('phone').filter(lambda x: len(x)>1)
grouped[['prefix','first','last','phone']]
```

```python
df.groupby('profession',as_index=False).agg(
    n=('profession','count')).query("profession=='Structural Engineer'").reset_index(drop=True)

```

```python
df.query("profession == 'Structural Engineer' & gender == 'Male'").groupby('profession', as_index=False).agg(n=('profession', 'count'))
```

```python
df.query("profession == 'Structural Engineer' & gender == 'Female' & province=='AB'").groupby('profession', as_index=False).agg(n=('profession', 'count'))
```
